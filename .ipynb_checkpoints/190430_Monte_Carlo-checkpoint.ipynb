# Sampling Method

Given a series of uniform distribution between 0~1, how to generate arbitrary distribution set?  

1. calculate $F(x)$(CDF)
1. generate $\lambda$ from (0, 1)uniform distribution  
1. solve $F(x) = \lambda$

Or **Metropolisâ€“Hastings** in MCMC

1. choose random number $x$ on domain of $f(x)$ 
1. compare $x$ with $f(x)/M$
1. if $x$ is greater, accept; if $x$ is smaller, reject.

# Nuclear Decay 

```python
N_Tl = 1000
N_Pb = 0
dt = 1
tau = 3.053 * 60
p = 1 - 2**(-dt / tau) # Decay probobility in dt
t = 0
t_end = 1000
_t = np.arange(0, t_end, dt)
decay_number = np.zeros_like(_t, dtype="int")

for i, j in enumerate(_t):
    decay_proposal = np.random.random(size=_t.size-np.sum(decay_number)) # For survival atoms
    decay_number[i] = np.sum(decay_proposal < p)
    
plt.plot(_t, N_Pb + np.cumsum(decay_number), label='Pb')
plt.plot(_t, N_Tl - np.cumsum(decay_number), label='Tl')
plt.legend()
```

# Brownian Motion



```python
# Initial position (0, 0)
t = 0
dt = 1
t_end = 100
t_size = (t_end - t)//dt
_t = np.arange(t, t_end, dt)
particle_n = 10
particles = np.zeros(particle_n, dtype='int') # 1-D random walking
history = np.zeros((t_size, particle_n))

for i, j in enumerate(_t):
    walking = np.random.choice([-1, 1], size=particle_n)
    particles = particles + walking
    history[i] = particles
    t += dt

plt.plot(_t, np.mean(history**2, axis=1)**0.5, 'p', alpha=.4);

plt.plot(_t, _t**0.5, '--', alpha=.4);
```

# Definite Integral

Sampling from a multidimensional domain, calculate integrand value(use $n-1$ dimensions)
at every sample point, compare random value and  integrand value(use $n$-th dimension) to
decide whether to accept it. Finally, calculate ratio as probability.  
Sampling scheme is often problem-specific.

```python
def integrand(x):
    return np.sin(x * (x-2)) + 1


x0, x1 = 0, 10
sample_n = 100000
sample_x = np.random.random(size=sample_n) * (x1 - x0) + x0  # Uniform sampling
sample_y = np.random.random(size=sample_n) * 2
mask = sample_y < integrand(sample_x)
below = np.sum(mask)
area = (x1 - x0) * 2
result = below / sample_n * area 
x = np.linspace(0, 10, 1000)
print(result)

plt.plot(x, integrand(x))
plt.plot(sample_x[mask], sample_y[mask], 'ro')
plt.plot(sample_x[np.logical_not(mask)], sample_y[np.logical_not(mask)], 'bp', alpha=.2)
```

```python
def integrand(x):
    return np.sin(1 / (x * (2 - x)))**2


x0, x1 = 0.01, 1.99
sample_n = 100000
sample_x = np.random.random(size=sample_n) * (x1 - x0) + x0  # Uniform sampling
mean_y = np.mean(integrand(sample_x))

result = mean_y * (x1 - x0) 
x = np.linspace(x0, x1, 1000)
print(result)

plt.plot(x, integrand(x))
```

# Maxwell Speed Distribution

```python
def collision(v1, v2):
    """
    Simulate collision of two molecules
    with speed v1, v2, respectively.
    
    return new speed of two molecules 
    """
    alpha = np.random.random() * 2 * np.pi
    gamma = np.random.random() * 2 * np.pi   
    _v1 = 0.5 * (v1**2 + v2**2) \
                + (0.5 * (v1**2 - v2**2) * np.cos(gamma)
                   + 0.5 * v1 * v2 * np.sin(alpha) * np.sin(gamma))
    _v2 = 0.5 * (v1**2 + v2**2) \
                - (0.5 * (v1**2 - v2**2) * np.cos(gamma)
                   + 0.5 * v1 * v2 * np.sin(alpha) * np.sin(gamma))
    return _v1**0.5, _v2**0.5


ctimes = 10**6
particle_num = 10**5
v_mean = 10
particle = np.random.randn(particle_num) + v_mean


for _ in range(ctimes):
    # Randomly select two molecules 
    n1, n2 = np.random.randint(0, particle.size, size=2)
    particle[[n1, n2]] = collision(*particle[[n1, n2]])
    
plt.hist(particle, bins=100);
```

# Ising Model

```python
%matplotlib notebook
n = 10
mag = np.ones((n, n))
J = 1.
k = 1

total_mag = np.sum(mag)
total_energy = 0
for i, row in enumerate(mag):
    for j, atom in enumerate(row):
        # Only include right and upper adjacents
        total_energy += - J * atom * (mag[(i + 1) % n, j] + mag[i, (j + 1) % n])

temp = 0.1
dT = 1e-1
temp_end = 5
equilibrium_n = 1000
energys = [total_energy]
mags = [total_mag]

def frame_iter():
    global temp, mag, total_energy, total_mag
    while temp < temp_end:
        for _ in range(equilibrium_n):
            # Ensure after n loops, system has been in equilibrium state
            for i in range(n):
                for j in range(n):
                    dE = J * 2 * mag[i,j] * (mag[(i + 1) % n, j] + mag[i, (j + 1) % n]\
                                            mag[(i - 1) % n, j] + mag[i, (j - 1) % n])
                    if dE < 0 or np.random.random() < np.exp(-dE / k / temp):
                        mag[i, j] = - mag[i, j]
                        total_energy += dE
                        total_mag += 2 * mag[i, j]

        energys.append(total_energy)
        mags.append(total_mag)
        temp += dT
        img.set_data(mag)
        ax.set_title(f"{temp}")
        yield img,


fig, ax = plt.subplots()
img = ax.imshow(mag)
img.set_clim(vmin=-1, vmax=1)
# line, = ax.plot([], [], 'ro')
_frames = frame_iter()
__frames = lambda i: next(_frames)
ani = animation.FuncAnimation(fig, __frames, frames=100, blit=True, interval=10)
```
