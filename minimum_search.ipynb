# Non-linear circuit 


For node V1, 
\begin{equation}
    \frac{V_1 - V_+}{R_1} + \frac{V_1}{R_2} + I_0\left[e^{(V_1 - V_2)/V_T} - 1\right] = 0
\end{equation}
For node V2, 
\begin{equation}
    \frac{V_+ - V_2}{R_3} - \frac{V2}{R_4} + I_0\left[e^{(V_1 - V_2)/V_T} - 1\right] = 0
\end{equation}

```python
V1, V2 = 0, 0
V0, R1, R2, R3, R4, I0, VT = 5, 1E3, 4E3, 3E3, 2E3, 3E-9, 0.05


def eq_arr(x):
    # Define equation function as 1-D array
    x1, x2 = x
    y1 = (x1 - V0) / R1 + x1/R2 + I0 * (np.exp((x1 - x2)/VT) - 1)
    y2 = (V0 - x2) / R3 - x2/R4 + I0 * (np.exp((x1 - x2)/VT) - 1)
    return np.array([y1, y2])


def eq_diff(x):
    x1, x2 = x
    y1_dx1 = 1/R1 + 1/R2 + I0 * np.exp((x1 - x2)/VT) * 1/VT
    y1_dx2 = I0 * np.exp((x1 - x2)/VT) * (-1/VT)
    y2_dx1 = I0 * np.exp((x1 - x2)/VT) * 1/VT
    y2_dx2 = -1/R3 -1/R4 + I0 * np.exp((x1 - x2)/VT) * (-1/VT)
    return np.array([[y1_dx1, y1_dx2], [y2_dx1, y2_dx2]])


volts = np.array([1., 1])
tol = 1E-6
err = 1
while abs(err) > tol**2:
    dv = np.linalg.solve(eq_diff(volts), eq_arr(volts))
    volts = volts - dv
    err = sum(dv**2)
    
print(f"Root of given circuit: V1={volts[0]:.4F}, V2={volts[1]:.4F}")
```

# Temperature of Lamb

```python
from scipy.constants import h, c, Boltzmann
from gaussxw import gaussxw


def lamb_radiation(temperature, l0, l1):
    """
    temperature: lamb temperature Unit: K
    l0: lower limit of wavelength
    l1: upper limit of wavelength

    return: ratio of a range integral over infinity integral
    """
    # Use Gauss integral method 
    # For computation convenience, set all constant 
    # coefficient to 1, which doesn't affect ratio value
    # Replace hc/(lambda k_ B T) with x 
    x1, x0 = h * c / (np.array([l0, l1]) * Boltzmann * temperature)
    integral = lambda x: x**5 / (np.exp(x) - 1)
    range_int = gauss_int(integral, x0, x1)
    integral_inf = lambda x:  1 / (1 - x)**2 * integral(x / (1 - x))
    inf_int = gauss_int(integral_inf, 0, 1, 100)
    
    return range_int / inf_int  


def gauss_int(f, x0, x1, sample_n=100): 
    x, w = gaussxw(sample_n)
    xp = 0.5 * (x1 - x0) * x + 0.5 * (x1 + x0)
    wp = 0.5 * (x1 - x0) * w
    
    return sum(f(xp) * wp)
    
    
def golden_search(f, a, b, tol):
    z = 1.618 
    c = b - (b - a)/z
    d = a + (b - a)/z
    
    fa, fb, fc, fd = f(a), f(b), f(c), f(d) 
    while abs(b - a) > tol:
        if fc > fd:
            # Update left node
            a, fa = c, fc
            c, fc = d, fd
            d = a + (b - a)/z
            fd = f(d)
        else:
            # Update left node
            b, fb = d, fd
            d, fd = c, fc
            c = b - (b - a)/z
            fc = f(c)
    return (b + a)/2

    
temp = np.linspace(300, 10E3)
wl1, wl2 = 390E-9, 750E-9
ratio = np.array([lamb_radiation(i, wl1, wl2) for i in temp])

# Ratio has maximum, find mimimum of its negative function 
negative_lamb = lambda x: -1 * lamb_radiation(x, wl1, wl2)
# From a rough scratch, maxima is in 4000~6000K
T0 = golden_search(negative_lamb, 4E3, 6E3, 1E-6)

ax = plt.gca()
ax.plot(temp, ratio)
ax.plot(T0, lamb_radiation(T0, wl1, wl2), marker="o", markersize=14)
ax.set_xlabel("Temperature/K")
ax.set_ylabel("Luminous Efficiency")

print(f"Luminous efficiency gets maxima at {T0:.2F}K")
```

Though, at 4529K, the luminous efficiency abtains its maxima,
the temperature is so high that it's greater than the melting 
point of tungsten. 

# Molecule potential minimum

```python
def molecule_potential(r, sigma=1.):
    return 1 * ((sigma/r)**6 - np.exp(-r/sigma))


def mp_diff(x, sigma=1.):
    return -6 * sigma**6 / x**7 + np.exp(-x/sigma) / sigma


def gradient_descent(df, x0, gamma, tol=1e-6):
    """
    df: derivative of function to be searched
    x0: initial guess
    gamma: adjustment speed
    tol: final precision demand
    
    return minimum of function 
    """
    x1 = x0 + 10 * tol
    while abs(x1 - x0) > tol:
        x0, x1 = x1, x0 - gamma * df(x0)
        
    return (x1 + x0) / 2
    

x = np.linspace(1, 20, 200)
x0 = gradient_descent(mp_diff, 1, 0.3, 1e-6)
print(f"Catch minimum value at {x0:.6F}nm")

ax = plt.gca()
ax.plot(x, molecule_potential(x))
ax.plot(x0, molecule_potential(x0), marker="o", markersize=10)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_position(("data", 0))
ax.set_xlim(0, )
```

# Simulated annealing 

```python
%matplotlib notebook
import matplotlib.animation as animation 


def annealing_ani(f, proposal, T0, cooling=0.95):
    """
    Demonstrate the annealing process with animation function
    
    i: frame iteration variable 
    f: function to be minimized
    proposal: iterator that generates next candidate
    T0: initial temperature
    cooling: cooling coefficient
    
    return: an iterator update line according to function call counts
    """
    # Initialize first parameters
    x = next(proposal)
    annealing_ani.call_counts = 0
    annealing_ani.T = T0
    annealing_ani.f = f(x)

    accept = True
    while accept:
        accept = False
        for _ in range(50 * len(x)):
            candidate = next(proposal)
            f_next = f(candidate)
            annealing_ani.call_counts += 1
            # Correct direction, accept it and update values
            if f_next <= annealing_ani.f:
                accept = True
                x = candidate
                annealing_ani.f = f_next
            # Probably wrong direction, accept with a probability
            elif np.exp((annealing_ani.f - f_next) / annealing_ani.T) > np.random.rand():
                accept = True
                x = candidate
                annealing_ani.f = f_next
        annealing_ani.T *= cooling
        # Update frame every 10 calls
        if annealing_ani.call_counts % 2000 == 0:
            head_tail = np.concatenate((x, x[0].reshape((1,-1))), axis=0)
            line.set_data(head_tail[:, 0], head_tail[:, 1])
            ax.set_title(f"T={annealing_ani.T:.4F}, "
                         f"distance={annealing_ani.f:.6F}")
            # Change normal function to iterator to store previous results
            yield line,


def neighbour(n=25):
    # Create n 2-D values between 0~1 as town's position
    np.random.seed(0)
    towns = np.random.rand(n, 2)
    while True:
        yield towns
        r1, r2 = np.random.choice(range(n), 2)
        new_towns = towns.copy()
        new_towns[[r1, r2]] = towns[[r2, r1]]
        towns = new_towns


def distance_2D(arr):
    dis = 0
    for i in range(len(arr) - 1):
        dis += (np.sum(arr[i + 1] - arr[i]) ** 2) ** 0.5
    dis += np.sum((arr[0] - arr[-1]) ** 2) ** 0.5

    return dis


# Create 25 towns position as a 2-D array
arr = neighbour(40)

fig, ax = plt.subplots()
# Fix x and y lim for animation
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
line, = ax.plot([], [], marker='o', markersize=5)

frame_iterator = annealing_ani(distance_2D, arr, 5*distance_2D(next(arr)))
# Define a frame-independent function for each frame call
frame_update = lambda i: next(frame_iterator)
ani = animation.FuncAnimation(fig, frame_update, frames=600, 
                              interval=400, blit=True, repeat_delay=3000)
ani.save("annealing_40towns.gif", writer="imagemagick", fps=20)
```
