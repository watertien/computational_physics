# Error comparision of forward difference and central difference 

Approximation error of forward difference:
\begin{equation}
    e_f = 2 \epsilon \frac{\left|\,f(x)\right|}{h} + \frac{1}{2}h \left|\,f''(x)\right|
\end{equation} 

Approximation error of central difference:
\begin{equation}
        e_c = 2 \epsilon \frac{\left|\,f(x)\right|}{h} + \frac{1}{24}h^2 \left|\,f''(x)\right|
\end{equation}

Solve inequality $e_c > e_f$, 
$$
    h > 12
$$

# Optimized h for second-order difference

According to 
\begin{equation}
    f''(x) = \frac{f(x+h) -2f(x) +f(x-h)}{h^2} - \frac{1}{12}h^2 f^{(4)}(x)
\end{equation}   

Rounding error: $4 \epsilon \left|\,f(x)\right| /h^2$  
Approximation error: $\frac{1}{12}h^2 \left|\,f^{(4)}(x)\right|$  
Total error:
$$
    e_2 = 4 \epsilon \frac{\left|\,f(x)\right|}{h^2} + \frac{1}{12}h^2 \left|\,f^{(4)}(x)\right| \ge \frac{1}{6} h^2 \left|\,f^{(4)}(x)\right| \approx 10^{-8}
$$
The inequality attains its minimum when and only when $h= \left(48 \epsilon \left|\,\frac{f(x)}{f^{(4)}(x)}\right| \right)^{1/4}$

# Integral of velocity

```python
def trape_int(f, x0, x1, n):
    # Check if n is greater than one to avoid zero division
    if n <= 1 or not isinstance(n, np.int64):
        raise ValueError("n should be an integer greater than 1")
    
    # Discretize x values
    x = np.linspace(x0, x1, n)
    # Calculate step value
    h = (x1 - x0) / (n - 1)
    # Trapezoidal integration formula
    return h * (1/2 * f(x0) + 1/2 * f(x1) 
            + np.sum([f(x0 + i * h) for i in range(1, n)]))


# Load data with names column and use time(column 0) as index
data = pd.read_table("data/week4_velocities.txt", index_col=0, 
                     names=['velocity'], header=None)
# Define a lambda function for dicrete time input
t2v = lambda t: data.loc[t, 'velocity']

# Cause raw data is discrete, h is not arbitrary
distances = [trape_int(t2v, data.index[0], i, 
                       i - data.index[0] + 1)
            for i in data.index[1:]]

# Plot velocity vs time
ax_v = plt.gca()
line_v, = ax_v.plot(data.index, data.velocity, 'b-', label="V(t)")
ax_v.set_xlabel("Time")
ax_v.set_ylabel("Velocity")
ax_v.set_xlim(0,)
ax_v.set_title("Velocity and Distance ")

# Plot distance vs time with sharing x axis with ax_v
ax_x = ax_v.twinx()
line_x, = ax_x.plot(data.index[1:], distances, 'r--', label="L(t)")
ax_x.set_ylabel("Distance")

lines = [line_x, line_v]
ax_x.legend(lines, [line.get_label() for line in lines]);
```

# Simpson rule vs trapezoidal rule

```python
from itertools import cycle


def simpson_int(f, x0, x1, n):
    # In order to divide x into even number strips,
    # n should be N_strip + 1
    if n % 2 == 0:
        raise ValueError("n should be an odd integer")
        return None

    x = np.linspace(x0, x1, n)
    # Step of x
    h = (x1 - x0) / (n - 1)
    # Different coefficient of odd/even terms
    # Use cycle iter to generate coefficients
    coe = cycle([4, 2])
    # Count from 1 to n-2
    # x[1:-1] is all elements in x except head and tail
    return h / 3 * (f(x0) + f(x1) + sum([f(i) * next(coe) for i in x[1:-1]]))
    

def trape_int(f, x0, x1, n):
    x = np.linspace(x0, x1, n)
    h = (x1 - x0) / (n - 1)
    return h * (1/2 * f(x0) + 1/2 * f(x1) 
            + sum([f(x0 + i * h) for i in range(1, n)]))


def x4_poly(x):
    return x**4 - 2 * x + 1


ana_int_func = lambda x: 1/5 * x**5 - x**2 + x 
ana_int = ana_int_func(2) - ana_int_func(0)

# Compare trapezoidal rule and Simpson rule for N=10, 100, 1000
for i in np.array([10, 100, 1000]) + 1:
    tra_int = trape_int(x4_poly, 0, 2, i)
    sim_int = simpson_int(x4_poly, 0, 2, i)
    print(f"relative error for N={i-1}:\n\t" 
          f"Simpson={(sim_int - ana_int) / ana_int :E}\n\t"
          f"Trapezoidal={(tra_int - ana_int) / ana_int :E}")

# Different steps for integration function  
ns = np.arange(10, 10**4+11, 100) + 1
tra_ints = np.array([trape_int(x4_poly, 0, 2, n) for n in ns])
sim_ints = np.array([simpson_int(x4_poly, 0, 2, n) for n in ns])
# Calculate error
tra_errors = (tra_ints - ana_int) / ana_int
sim_errors = (sim_ints - ana_int) / ana_int

# Plot N vs error for two methods
ax = plt.gca()
lines = plt.plot(ns, tra_errors, ns, sim_errors)
ax.set_xlabel("Strip number of x")
ax.set_ylabel("Error")
ax.set_title("Error Comparsion")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlim(10,)
ax.legend(lines, ["trapezoidal rule", "Simpson rule"]);
```

From above figure, there is a strong log-linear correlation between 
the error and the strip number for both integration methods, which means
the error and strip number have a power-law relation, 
e.g. $\textbf{error} = a\cdot x^{-k} (k>0)$. 
In particular, Simpson rule has significant advantages
over trapezoidal rule with lower error and more sensitive 
to the strip number. In addition, for Simpson rule, when strip number increases
to around 4000, the difference between analytical result 
and numerical result is close to Python float rounding error, 
which proves the excellence of Simpson rule in another perspective.

# Bessel function of the first kind

```python
def functionJ(m, x):
    # Define integrand by m and x 
    integrand = lambda theta: np.cos(m * theta - x * np.sin(theta))
    # Use Simpson rule to calculate m-th order Bessel function of the first kind
    # Strip number n
    n = 10**3 + 1
    return 1 / np.pi * simpson_int(integrand, 0, np.pi, n)


x = np.linspace(0, 20, 100)
jxs = np.array([functionJ(i, x) for i in range(3)])

# Plot J0, J1, J2, x from 0 to 20
colors = ["black", "red", "blue"]
ax = plt.gca()
# Flatten a list of list to share x value
plot_data = [j for i in zip([x] * len(jxs), jxs) for j in i ]
# Unpack plot_data to plot three lines
lines = ax.plot(*plot_data)
for i, line in enumerate(lines):
    line.set_color(colors[i])
    line.set_label(f"$J_{i}(x)$")
ax.set_xlabel("x")
ax.set_xlim(0,)
ax.legend()
```

## Airy disk

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm


def airy_disk(wavelength, r):
    k = 2 * np.pi / wavelength
    _disk = np.zeros(r.shape)
    # Set std error to ignore ZeroDivisionError
    np.seterr(divide="ignore")
    # Use np.divide to avoid ZeroDivisionError
    _disk = np.divide(functionJ(1, k * r), k * r)**2
    # J1(x)/x = 1/2 when x gets close to 0
    _disk[r == 0] = 0.5
    return _disk


# Length unit: um
x = np.linspace(-1, 1+.1, 400)
y = np.linspace(-1, 1+.1, 400)
wavelength = 0.5 # 500nm

xx, yy = np.meshgrid(x, y)
r = (xx**2 + yy**2)**0.5
zz = airy_disk(wavelength, r)

fig = plt.figure(figsize=(10, 5))

# Set a suitable color for 2-D and 3-D
color_name = "gnuplot"
ax2d = fig.add_subplot(121)
img = ax2d.imshow(zz, extent=(-1, 1, -1, 1), cmap=color_name,
                  aspect="auto", origin='lower')
ax2d.set_xlabel("x/um")
ax2d.set_ylabel("y/um")
ax2d.set_title("Airy Disk in 2-D")

ax3d = fig.add_subplot(122, projection='3d')
surface = ax3d.plot_surface(xx, yy, zz, alpha=.4, cmap=color_name)
ax3d.set_xlabel("x/um")
ax3d.set_ylabel("y/um")
ax3d.set_zlabel("I(r)")
ax3d.set_title("Airy Disk in 3-D")

cmp = fig.colorbar(img, ax=ax3d)
fig.tight_layout()
fig.savefig("airy_disk.pdf")
```

# Adaptive integration

```python
def ada_int(integrand, method, order, x0, x1, tol):
    '''adaptive integral of given method
    
    integrand: function to be integrated
    order: error order of given method
    x0: lower limit of integration
    x1: upper limit of integration 
    tol: demanding error limit of integral 
    
    return integral result, iteration times, final error
    '''
    if method == "trap":
        ada_sub = ada_trap_sub
    elif method == "simp":
        ada_sub = ada_simp_sub 
    else:
        raise ValueError("Only support 'trap' or 'simp'")
    
    ada_result = ada_sub(integrand, x0, x1)
    
    last_result, n = next(ada_result)
    current_result, n = next(ada_result)

    # Calculate first error value
    error = abs((current_result - last_result) / (2 * 2**order - 1))
    
    print(f"Strip number N: {n:6}  "
          f"Integral result: {current_result}  "
          f"Estimated error: {error}")
          
    while error > tol:
        last_result = current_result
        current_result, n = next(ada_result)
        error = abs((current_result - last_result) / (2 * 2**order - 1))
        print(f"Strip number N: {n:6}  "
              f"Integral result: {current_result}  "
              f"Estimated error: {error}")
    
    return current_result, n, error

           
def ada_trap_sub(f, x0, x1, n=1):
    '''
    a generator function to compute S2 adaptively
    '''
    prev_int = trape_int(f, x0, x1, n)
    
    while True:
        h = (x1 - x0) / (n)
        # Change of 2N from N
        delta = sum([f(x0 + i * h) for i in range(1, n, 2)])
        # Adaptive trape int formula
        next_int = 1 / 2 * prev_int + h * delta
        yield next_int, n
        # For next iteration 
        prev_int = next_int
        # Update n value
        n *= 2 


def ada_simp_sub(f, x0, x1, n=2):
    '''
    a generator function to compute S2 adaptively
    '''
    prev_int = simpson_int(f, x0, x1, n)

    h = (x1 - x0) / (n)
    # Initialize I1 and T1
    part_1 = 1 / 3 * (f(x0) + f(x1) 
             + 2 * sum([f(x0 + i * h) for i in range(2, n-1, 2)]))
    part_2 = 0

    while True:
        # Change of 2N from N
        h = (x1 - x0) / (n)
        part_1 = part_1 + part_2 # I2 = I1 + T1 
        part_2 = 2 / 3 * sum([f(x0 + i * h) for i in range(1, n, 2)]) # T2
        next_int = h * (part_1 + 2 * part_2)
        yield next_int, n
        # For next iteration 
        prev_int = next_int
        # Update n value
        n *= 2 

              
def trape_int(f, x0, x1, n):
    x = np.linspace(x0, x1, n+1)
    h = (x1 - x0) / (n)
    return h * (1/2 * f(x0) + 1/2 * f(x1) 
            + sum([f(x0 + i * h) for i in range(1, n+1)]))        


def simpson_int(f, x0, x1, n):
    # In order to divide x into even number strips,
    # n should be N_strip, namely an even number
    if n % 2 == 1:
        raise ValueError("n should be an odd integer")
        return None

    x = np.linspace(x0, x1, n+1)
    # Step of x
    h = (x1 - x0) / (n)
    # Different coefficient of odd/even terms
    # Use cycle iter to generate coefficients
    coe = cycle([4, 2])
    # Count from 1 to n-2
    # x[1:-1] is all elements in x except head and tail
    return h / 3 * (f(x0) + f(x1) + sum([f(i) * next(coe) for i in x[1:-1]]))
    

def sin_square(x):
    return np.sin((100 * x)**0.5)**2


print("Adaptive trapezoidal integral:")
ada_int(sin_square, "trap", 1, 0, 1, 1e-6)
print("Adaptive Simpson integral:")
ada_int(sin_square, "simp", 3, 0, 1, 1e-6);
```

# Blackbody radiation 

\begin{equation}
    I(\omega) = \frac{\hbar}{4 \pi^2 c^2}\, \frac{\omega^3}{e^{\hbar \omega/(k_B T)} - 1}
\end{equation}
$I(\omega)$ is the radiation power per unit area in unit frequency range from an ideal blackbody.  

The total power of blackbody per unit area:  
\begin{equation}
        W = \int_{0}^{\infty} I(\omega) \textbf{d}\omega 
\end{equation}

   Substitute $\hbar \omega/(k_B T)$ for x, the above equation becomes  

$$
W = \frac{k_B^4 T^4}{4 \pi^2 c^2 \hbar^3} \int_0^{\infty} \frac{x^3}{e^x - 1} \textbf{d}x 
$$

```python
from gaussxw import gaussxw


def blackbody_radiation_toinf(x):
    np.seterr(over="print")
    f = lambda a: a**3 / (np.e**a - 1)
    return 1 / (1 - x)**2 * f(x / (1 - x))


def gauss_int(f, x0, x1, sample_n=100): 
    x, w = gaussxw(sample_n)
    xp = 0.5 * (x1 - x0) * x + 0.5 * (x1 + x0)
    wp = 0.5 * (x1 - x0) * w
    
    return sum(f(xp) * wp)


stefan_int = gauss_int(blackbody_radiation_toinf, 0, 1, 100)
print(stefan_int)
```

From above
$$
    \int_0^{\infty} \frac{x^3}{e^x - 1} \textbf{d}x \approx 6.49394
$$
And according to Stefan's law, 
$$
    \sigma = 6.49394 \cdot \frac{k_B^4}{4 \pi^2 c^2 \hbar^3}
$$

```python
from scipy.constants import c, hbar, k, Stefan_Boltzmann

num_sigma = stefan_int * k**4 / (4 * np.pi**2 * c**2 * hbar**3)

print(f"Numerical intgral result for sigma: {num_sigma}\n"
      f"Scipy constant for sigma: {Stefan_Boltzmann}\n"
      f"Relative error: {(num_sigma - Stefan_Boltzmann)/Stefan_Boltzmann}")
```

# Gravity of a uniform plate

For a tiny piece of plate, the force along z axis of this piece is:
$$
    \textbf{d}f = G \sigma \frac{\textbf{d}x\textbf{d}y}{(x^2 + y^2 + z^2)} \cos{\theta}
$$
in which $\cos{\theta} = z / (x^2 + y^2 + z^2)^{1/2}$. Therefore, the total force of the plate
$$
    F = G \sigma z \iint_{S}\frac{\textbf{d}x\textbf{d}y}{(x^2 + y^2 + z^2)^(3/2)}
$$

```python
from scipy.constants import G


def gauss2D_int(f, x0, x1, y0, y1, z, n_x=100, n_y=100):
    '''Function to calculate double integral over a rectangular area
    f: integrand function
    x0: lower limit of x
    x1: upper limit of x
    y0: lower limit of y
    y1: upper limit of y
    z: z value of the point
    n_x: sample number on x axis
    n_y: sample number on y axis
    '''
    x, wx = gaussxw(n_x)
    x_ex = 0.5 * (x1 - x0) * x + 0.5 * (x1 + x0)
    wx_ex = 0.5 * (x1 - x0) * wx
    
    # Check if we can avoid some repeat calculation 
    if n_y != n_x:
        y, wy = gaussxw(n_y)
    else:
        y, wy = x, wx

    y_ex = 0.5 * (y1 - y0) * x + 0.5 * (y1 + y0)
    wy_ex = 0.5 * (y1 - y0) * wy
    
    xx, yy = np.meshgrid(x_ex, y_ex)
    # Use broadcast to create a weight mesh 
    # Reshape wy_ex as a column vector
    wy_ex = wy_ex.reshape((-1, 1))
    wxy = wx_ex * wy_ex
    
    # addition along all axes
    return np.sum(f(xx, yy, z) * wxy)


def gravity_int(x, y, z):
    return (x**2 + y**2 + z**2)**(-3 / 2)


mass_sigma = 10**3 / (10 * 10) # Mass density of the plate Unit: kg/m^2
# Generate test points
z = np.linspace(0, 10)
f_z = G * mass_sigma * np.array([i * gauss2D_int(gravity_int, -5, 5, -5, 5, i) for i in z])

plt.figure(figsize=(15, 10))

plt.subplot(211)
plt.plot(z, f_z, marker='o')
plt.xlabel("z")
plt.ylabel("$F_z/N$")
plt.xlim(0, )
plt.ylim(0, )
plt.title("Gravity along z axis from a square plate")

# Create a densy array near z=0 
dense_z = np.linspace(0, 0.2, 100)

# Concatenate two arrays to get a better z array from 0 to 10
z_imporved = np.concatenate((dense_z, z[1:]))
f_z_improved = G * mass_sigma * np.array([i * gauss2D_int(gravity_int, -5, 5, -5, 5, i) for i in z_imporved])

plt.subplot(212)
plt.plot(z_imporved, f_z_improved, marker='o')
plt.xlabel("z")
plt.ylabel("$F_z/N$")
plt.xlim(0, )
plt.ylim(0, )
plt.title("Gravity along z axis from a square plate(with more samples around 0)");
```
